```
# Declare a fuction ret_element that return indexed element from input vector
ret_element <- function(input_vector, index) {
  return (input_vector[index])
}

input_vector <- c("Sun","Mon","Tue","Wed","Thurs","Fri","Sat")
index <- 3

ret_element(input_vector, index)
```

```
# Redefine above function  to have default argument index to have value 0 such that it will return first element if second argument is missing
ret_element <- function(input_vector, index=1) {
  return (input_vector[index])
}

input_vector <- c("Sun","Mon","Tue","Wed","Thurs","Fri","Sat")


# Second argument not missing
ret_element(input_vector, 3)

# Second argument missing
ret_element(input_vector)
```

```
#using above function define a function that will extract an element in a matrix
#[1 2 3   1 2  -->   ret_eement ([2 5] , 1) ] -->   2   
# 4 5 6

MyFunction <- function(mat, row, col) {
  return(ret_element(mat[, col], row))
}

mat <- matrix(1:6, nrow=2, ncol=3, byrow=TRUE)

MyFunction(mat, 1, 2)
```

```
# Write a user defined function create_frame that takes these vecotrs as inputs and represents them into a data frame 
# with x1 labelled as age and x2 labelled as Systolic Blood Pressure 

create_frame <- function(x1, x2) {
  frame <- data.frame(
    age = x1,
    systolic_blood_pressure = x2
  )
  frame[] <- lapply(frame, as.integer)
  return(frame)
}
create_frame(x1, x2)
```

```
# write a R function that inputs a number and return double of it if it is positive
#returns half if it is negative and return a call to same function with number+1 
#uniform random number in range -2  and 2 

MyFunction <- function(number) {
  if(number > 0) {
    return(number * 2)
  } else {
    return(MyFunction(((number / 2) + 1)))
  }
}

number <- runif(1, -2, 2)
number
MyFunction(number)
```

```
# Write a R function that return all prime numbers between(and including) start and end parameter 
# where default value for start is 2 and end is 10
prime_numbers <- function(start=2, end=10) {
  if(end >= start) {
    x = seq(start, end)
    prime_nums = c()
    for(i in x) {
      if(any(x == i)) {
          prime_nums = c(prime_nums, i)
          x = c(x[(x %% i) != 0], i)
      }
    }
    return(prime_nums)
  }
} 
prime_numbers()
```

```
# show how u use this method to obtain first 10 prime numbers with out changing its code.
prime_numbers(2, 30)
```
